```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#plot target dist#target distribution
#sns.countplot(df.target,hue='sex',plaette=['green','red])
#sns.countplot(x=df.target,data=df,plaette=['#1CA53B','red'])
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
#seaborn.pair plot(df,hue='size')
plt.xticks([0,1],labels=['not disease','disease'])
plt.title('target distribution')
```

```python
#plot target dist#target distribution
#sns.countplot(df.target,hue='sex',plaette=['green','red])
#sns.countplot(x=df.target,data=df,plaette=['#1CA53B','red'])
sns.countplot(x=df.target,data=df,palette=['#1CA53B','red'])
#seaborn.pair plot(df,hue='size')
plt.xticks([0,1],labels=['not disease','disease'])
plt.title('target distribution')
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['yellow','magenta'])
plt.legend(["haven't disease","have disease"])
plt.title("heart disease frequency for ages")
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='green')
sns.distplot(df.age[df['target']==1],bins=30,color='magenta')
plt.legend()
plt.title("heart disease frequency for ages")
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.displot(df.age[df['target']==0],bins=30,color='green')
sns.displot(df.age[df['target']==1],bins=30,color='magenta')
plt.legend(["haven't disease","have disease"])
plt.title("heart disease frequency for ages")
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```
